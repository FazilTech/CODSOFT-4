# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
sns.set()
```

# Import CSV File

```python
raw_data = pd.read_csv("advertising.csv")
raw_data.head()
```

# Check for NULL Values

```python
raw_data.isnull().sum()
```

```python
# no null values
# now describe the dataset
data = raw_data.copy()
data.describe(include='all')
```

# Train Test Split

```python
# TV has the more better data compared to other two 
# Taking TV as the input value for the prediction
inputs = data[['TV']]
target = data[['Sales']]

#split the dataset
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=365)
model = LinearRegression()
model.fit(x_train, y_train)
```

# Prediction for Train DataSet

```python
x_train_prediction = model.predict(x_train)
x_train_prediction
```

# Prediction for Test DataSet

```python
x_test_prediction  = model.predict(x_test)
x_test_prediction
```

```python
# plot the prediction data
y_hat = x_test_prediction
plt.scatter(y_test, y_hat)
plt.xlabel('Target')
plt.ylabel('Prediction')
```

# Calculate the Difference

```python
# create a new DataColumn for calculating the difference
df_pf = pd.DataFrame(y_hat, columns=['Predictions'])
df_pf['Target'] = y_test
df_pf
```

```python
# lot of missing values
# to take care of the missig values
y_test = y_test.reset_index(drop=True)
df_pf['Target'] = y_test
df_pf
```

```python
# calculating the difference%
df_pf['Residual'] = df_pf['Target'] - df_pf['Predictions']
df_pf['Difference %'] = np.absolute((df_pf['Residual']/df_pf['Target'])*100)
sort_data=df_pf.sort_values(by=['Difference %'])
sort_data
```
